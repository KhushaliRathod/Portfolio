Importing Dependencies

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection and Analysis

```python
diabetes_dataset = pd.read_csv('/diabetes.csv')
```

```python
diabetes_dataset.head()
```

```python
diabetes_dataset.shape
```

```python
diabetes_dataset['Outcome'].value_counts()
```

```python
diabetes_dataset.groupby('Outcome').mean()
```

0 = Non Diabetic
1 = Diabetic

```python
X = diabetes_dataset.drop(columns = 'Outcome', axis=1)
Y = diabetes_dataset['Outcome']
```

```python
X.head()
```

```python
Y.head()
```

```python
#Data Standardization
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
std_data = scaler.transform(X)
```

```python
print(std_data)
```

```python
X = std_data
```

```python
print(X)
print(Y)
```

```python
#Train test split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size =0.2,stratify=Y ,random_state =2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#training the model
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train,Y_train)
```

```python
#Accuracy score on the training data
#The first line generates predictions from the model based on the training data.
#The second line compares those predictions to the actual training labels and calculates the accuracy,
#showing how well the model fits the training data.
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of training data : ',training_data_accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of test data : ',test_data_accuracy)
```

```python
#Making prediction with any input
```

```python
inputdata = (10,168,74,0,0,38,0.537,34)
```

```python
inputdata_as_npa =np.asarray(inputdata)
```

```python
inputdata_reshape = inputdata_as_npa.reshape(1,-1)

stds_data = scaler.transform(inputdata_reshape)
print(stds_data)
predict = classifier.predict(stds_data)
print('Prediction is = ' , predict)
if predict[0] == 0:
    print('The person is not diabetic')
else:
    print('The person is diabetic')
```
